```python
import collections.abc
from tkinter import *
from experta import *
from PIL import Image, ImageTk

# ---------------------------------------------------------
collections.Mapping = collections.abc.Mapping
collections.MutableMapping = collections.abc.MutableMapping
collections.Iterable = collections.abc.Iterable
# ---------------------------------------------------------

class find_singer(KnowledgeEngine):
    @DefFacts()
    def _initial_action(self):
        self.gui_form1 = Tk()
        self.gui_form1.title("Find your singer")
        self.gui_form1.geometry("800x600")
        
        try:
            bg_image_data = Image.open("background.jpg")
            bg_image_data = bg_image_data.resize((800, 600), Image.Resampling.LANCZOS)
            self.bg_photo = ImageTk.PhotoImage(bg_image_data)
            
            bg_label = Label(self.gui_form1, image=self.bg_photo)
            bg_label.place(x=0, y=0, relwidth=1, relheight=1)
        except Exception as e:
            print(f"Background image not found or error: {e}")
            self.gui_form1.configure(bg="#f0f0f0")
        # -------------------------------------------

        lb0 = Label(self.gui_form1, text="Hi! I am here to help you choose your singer.\n", bg="#123456", fg="#fff", width="60")
        lb0.pack(pady=10)
        
        # Singer Statue
        lb1 = Label(self.gui_form1, text="Singer Statue (modern / old)")
        self.en1 = Entry(self.gui_form1)
        lb1.pack()
        self.en1.pack()
        
        # Century
        lb2 = Label(self.gui_form1, text="Century (21st / 20th)")
        self.en2 = Entry(self.gui_form1)
        lb2.pack()
        self.en2.pack()
        
        # Tempo
        lb4 = Label(self.gui_form1, text="Tempo (fast / slow)")
        self.en4 = Entry(self.gui_form1)
        lb4.pack()
        self.en4.pack()
        
        # Gender
        lb5 = Label(self.gui_form1, text="Gender (male / female)")
        self.en5 = Entry(self.gui_form1)
        lb5.pack()
        self.en5.pack()
        
        # Songs Type
        lb6 = Label(self.gui_form1, text="Songs Type (romance / diverse / sad)")
        self.en6 = Entry(self.gui_form1)
        lb6.pack()
        self.en6.pack()
        
        # Nationality
        lb3 = Label(self.gui_form1, text="Nationality (egyptian / american)")
        self.en3 = Entry(self.gui_form1)
        lb3.pack()
        self.en3.pack()

        btn = Button(self.gui_form1, command=self.get_values, text="Get Values & Find", bg="#123456", fg="#fff", font=("Arial", 10, "bold"))
        btn.pack(pady=20)

        self.gui_form1.mainloop()
        yield Fact(action="find_singer")

    def get_values(self):
        self.declare(Fact(singer_statue=self.en1.get().strip()))
        self.declare(Fact(century=self.en2.get().strip()))
        self.declare(Fact(tempo=self.en4.get().strip()))
        self.declare(Fact(gender=self.en5.get().strip()))
        self.declare(Fact(types=self.en6.get().strip()))
        self.declare(Fact(nationality=self.en3.get().strip()))
        self.gui_form1.destroy()

    @Rule(Fact(action='find_singer'), NOT(Fact(singer_statue=W())))
    def statue(self):
        self.declare(Fact(singer_statue=input(" Please enter 'singer_statue': [modern , old] : ")))

    @Rule(Fact(action='find_singer'), NOT(Fact(century=W())))
    def century(self):
        self.declare(Fact(century=input("Please enter the 'century' of the singer: [21st , 20th] :")))

    @Rule(Fact(action='find_singer'), NOT(Fact(tempo=W())))
    def tempo(self):
        self.declare(Fact(tempo=input("Please enter the 'tempo' common in his/her songs: [slow , fast] : ")))

    @Rule(Fact(action='find_singer'), NOT(Fact(gender=W())))
    def gender(self):
        self.declare(Fact(gender=input("Please enter his/her 'gender': [male, female] : ")))

    @Rule(Fact(action='find_singer'), NOT(Fact(types=W())))
    def types(self):
        self.declare(Fact(types=input("Please enter 'type' of his most songs: [sad, romance ,diverse] : ")))

    @Rule(Fact(action='find_singer'), NOT(Fact(nationality=W())))
    def nationality(self):
        self.declare(Fact(nationality=input("Please enter the 'nationality' of the singer: [egyptian, american] : ")))

    # ----------------- Egyptian Singers Rules -----------------
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_1(self):
        self.declare(Fact(singer="Mohamed Fouad"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_2(self):
        self.declare(Fact(singer="Mohamed Mounir"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_3(self):
        self.declare(Fact(singer="Samira Said"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_4(self):
        self.declare(Fact(singer="Sherine"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_5(self):
        self.declare(Fact(singer="Tamer Hosny"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_6(self):
        self.declare(Fact(singer="Amr Diab"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_7(self):
        self.declare(Fact(singer="Carmen Soliman"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_8(self):
        self.declare(Fact(singer="Amal Maher"))

    # Diverse
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_9(self):
        self.declare(Fact(singer="Hamada Helal"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_10(self):
        self.declare(Fact(singer="Mostafa Qamar "))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_11(self):
        self.declare(Fact(singer="Angham"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_12(self):
        self.declare(Fact(singer="Haifaa Wahby"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_13(self):
        self.declare(Fact(singer="Ahmed Saad"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_14(self):
        self.declare(Fact(singer="Hany Adel"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_15(self):
        self.declare(Fact(singer="Yasmin Aly"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_16(self):
        self.declare(Fact(singer="Dina Elwadedi"))

    # Sad
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_17(self):
        self.declare(Fact(singer="Hany Shaker"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_18(self):
        self.declare(Fact(singer="Mostafa Kamel"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_19(self):
        self.declare(Fact(singer="Aida Elayouby"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_20(self):
        self.declare(Fact(singer="Dalida"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_21(self):
        self.declare(Fact(singer="Muslim"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_22(self):
        self.declare(Fact(singer="Tamer Ashour"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_23(self):
        self.declare(Fact(singer="Asala"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_24(self):
        self.declare(Fact(singer="Yasmina"))

    # Egyptian 20th
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_25(self):
        self.declare(Fact(singer="Baligh Hamdy"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_26(self):
        self.declare(Fact(singer="Abdelhaleem Hafez"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_27(self):
        self.declare(Fact(singer="Soaad Hosny"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_28(self):
        self.declare(Fact(singer="Shadia"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_29(self):
        self.declare(Fact(singer="Ali Hemida"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_30(self):
        self.declare(Fact(singer="Ehab Tawfiq"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_31(self):
        self.declare(Fact(singer="Yousra"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="egyptian"))
    def singer_32(self):
        self.declare(Fact(singer="Nagat Elsagheera"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_33(self):
        self.declare(Fact(singer="Bob Azzam"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_34(self):
        self.declare(Fact(singer="Mohamed Abdelwahab")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_35(self):
        self.declare(Fact(singer="Ratiba Elhefny"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_36(self):
        self.declare(Fact(singer="Umm Kulthum"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_37(self):
        self.declare(Fact(singer="Hesham Abbas"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_38(self):
        self.declare(Fact(singer="Ali El7agar"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_39(self):
        self.declare(Fact(singer="Gawaher")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="egyptian"))
    def singer_40(self):
        self.declare(Fact(singer="Anoushka"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_41(self):
        self.declare(Fact(singer="Sayed Darwish"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_42(self):
        self.declare(Fact(singer="Salama Hegazy"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_43(self):
        self.declare(Fact(singer="Leila Mourad"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_44(self):
        self.declare(Fact(singer="Huda Sultan"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_45(self):
        self.declare(Fact(singer="Mohamed Mohie"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_46(self):
        self.declare(Fact(singer="Hamid Al Shaeri")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_47(self):
        self.declare(Fact(singer="Simone Philip Kamel"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="egyptian"))
    def singer_48(self):
        self.declare(Fact(singer="Gala Elhadedy"))

    # ----------------- American Singers Rules -----------------
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_49(self):
        self.declare(Fact(singer="Rascal Flatts"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_50(self):
        self.declare(Fact(singer="Keith Urban"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_51(self):
        self.declare(Fact(singer="Christina Aguilera"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_52(self):
        self.declare(Fact(singer="Vanessa Carlton"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_53(self):
        self.declare(Fact(singer="Sufjan Stevens"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_54(self):
        self.declare(Fact(singer="Bruno Mars"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_55(self):
        self.declare(Fact(singer="Billie Eilish"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_56(self):
        self.declare(Fact(singer="Rihanna"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_57(self):
        self.declare(Fact(singer="Afroman"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_58(self):
        self.declare(Fact(singer="Rob Beckley "))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_59(self):
        self.declare(Fact(singer="Alicia Keys"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_60(self):
        self.declare(Fact(singer="Adele"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_61(self):
        self.declare(Fact(singer="Adam Levine"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_62(self):
        self.declare(Fact(singer="Justin Babier"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_63(self):
        self.declare(Fact(singer="SIA"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_64(self):
        self.declare(Fact(singer="Lana Del Rey"))

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_65(self):
        self.declare(Fact(singer="Jason Isbell"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_66(self):
        self.declare(Fact(singer="Jesse Malin"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_67(self):
        self.declare(Fact(singer="Amy Winehouse"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_68(self):
        self.declare(Fact(singer="M.I.A"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_69(self):
        self.declare(Fact(singer="Eric R. Danton"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_70(self):
        self.declare(Fact(singer="Ellen Johnson"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_71(self):
        self.declare(Fact(singer="St. Vincent")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="21st"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_72(self):
        self.declare(Fact(singer="Robyn")) 

    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_73(self):
        self.declare(Fact(singer="Elvis Presley"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_74(self):
        self.declare(Fact(singer="Bruce Springsteen")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_75(self):
        self.declare(Fact(singer="Whitney Houston"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_76(self):
        self.declare(Fact(singer="Patsy Cline")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_77(self):
        self.declare(Fact(singer="Stevie Wonder")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_78(self):
        self.declare(Fact(singer="David Bowie")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_79(self):
        self.declare(Fact(singer="Madonna")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="romance"),Fact(nationality="american"))
    def singer_80(self):
        self.declare(Fact(singer="Reba McEntire")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_81(self):
        self.declare(Fact(singer="John Lennon")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_82(self):
        self.declare(Fact(singer="Frank Sinatra")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_83(self):
        self.declare(Fact(singer="Aretha Franklin"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_84(self):
        self.declare(Fact(singer="Bette Midler")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_85(self):
        self.declare(Fact(singer="Micheal Jackson"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_86(self):
        self.declare(Fact(singer="Johnny Cash"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_87(self):
        self.declare(Fact(singer="Cassandra Wilson"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="diverse"),Fact(nationality="american"))
    def singer_88(self):
        self.declare(Fact(singer="Celia Cruz"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_89(self):
        self.declare(Fact(singer="Prince"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_90(self):
        self.declare(Fact(singer="James Brown"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_91(self):
        self.declare(Fact(singer="Ann Hampton Callaway"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="old"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_92(self):
        self.declare(Fact(singer="Aretha Franklin")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_93(self):
        self.declare(Fact(singer="Stevie Wonder"))
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="male"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_94(self):
        self.declare(Fact(singer="Snoop Dogg")) 
        
    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="slow"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_95(self):
        self.declare(Fact(singer="Beyonce")) 

    @Rule(Fact(action='find_singer'),Fact(singer_statue="modern"),Fact(century="20th"),Fact(tempo="fast"),Fact(gender="female"),
          Fact(types="sad"),Fact(nationality="american"))
    def singer_96(self):
        self.declare(Fact(singer="Carly Simon"))  

    @Rule(Fact(action='find_singer'),Fact(singer=MATCH.singer))
    def singer(self, singer):
        id_singer = singer
        gui_form2 = Tk()
        gui_form2.title("Your singer Answer")
        gui_form2.geometry("600x500")
        gui_form2.configure(bg="#f0f0f0")
        
        Label(gui_form2, text="The most probable Singer is:", bg="#123456", fg="#fff", width="800").pack()
        
        try:
            image_name = f"photos/{id_singer}.jpg"
            img_data = Image.open(image_name)
            
            img_data = img_data.resize((250, 250), Image.Resampling.LANCZOS)
            photo = ImageTk.PhotoImage(img_data)
            
            img_label = Label(gui_form2, image=photo)
            img_label.image = photo 
            img_label.pack(pady=20)
        except Exception as e:
            print(f"Image for {id_singer} not found: {e}")
            Label(gui_form2, text="(Image not available)", fg="red", bg="#f0f0f0").pack(pady=10)

        var = StringVar()
        var.set(id_singer)
        Label(gui_form2, textvariable=var, bg="#678954", fg="#fff", font=("Arial", 16, "bold"), width="40").pack(pady=20)
        
        print("The most probable Singer you search is %s\n" %(id_singer))
        gui_form2.mainloop()

if __name__ == "__main__":
    engine = find_singer()
    engine.reset()
    engine.run()
```
